#### effective mass function

```python
def effective_mass(propagator):
    len_p=len(propagator)
    output=len_p*[None]
    for i in range(len_p/2-1):
        output[i]=(log(propagator[i]/propagator[i+1])).real
    for i in range(-len_p/2+1,0):
        output[i]=(log(propagator[i+1]/propagator[i])).real
    
    return output
```

```python
# Two subplots, the axes array is 1-d
time_slices=range(t_sites)
num_particles=len(meson_propagators_all)
num_configs=len(meson_propagators_all[0])

f, axarr = plt.subplots(nrows=num_particles, ncols=num_configs, sharex=True,sharey=True,figsize=(12,12))

for part in range(num_particles):
    for conf in range(num_configs):
        for i in range(len(meson_propagators_all[0][0])):
            axarr[part,conf].scatter(time_slices,effective_mass(meson_propagators_all[part][conf][i]))
   
```
