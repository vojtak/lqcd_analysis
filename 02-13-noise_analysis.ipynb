## set parameters of the analysis 

#### path to folder with results

```python
path_results  ="../X_VOJTA_code/run_x64/results/"
```

#### type of measurement (aka measurement folder name)

```python
measurements_noise=["nn_01_correlators_two_hardons",
                    "nn_04_correlators_two_hardons",
                    "nn_10_correlators_two_hardons",
                    "nn_20_correlators_two_hardons",
                    "nn_50_correlators_two_hardons"]
measurement_full  =["nn_01_correlators_two_hardons"]
```

#### particle(s)  (in list)

```python
particles_noise=["pion-sigma_tree"]
particles_full =["pion-sigma_tree_NOnoise"]
```

## extracting data

```python
propagators_noise=[extract_propagators(path_results,measurement,particles_noise) for measurement in measurements_noise]
propagators_full =[extract_propagators(path_results,measurement,particles_full ) for measurement in measurement_full  ]
```

## plot of differences in propagator values

```python
fig, ax = subplots(figsize=(30,20))
noises=["$1$","$4$","$10$","$20$","$50$"]
colors=["red","red","green","green","green"]
sizes=[200,200,300,300,300]

conf=0
t_slice=2


for t_s in range(24):

    subplot(3,24,t_s+1)

    if(t_s==0):
        ylabel('real part [0.5x,1.5x]',fontsize=24, color='green')
    
    for ii,prop in enumerate(propagators_noise):

        to_plot=[(x.real) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot,color=colors[ii],marker=noises[ii],s=sizes[ii])

    for prop in propagators_full:
 
        to_plot=[(x.real) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot, s=200)

    ylim([0.5*to_plot[0],1.5*to_plot[0]])
    xticks([])
    title("t="+str(t_s),fontsize=24, color='green')
    
    subplot(3,24,t_s+25)

    if(t_s==0):
        ylabel('imag part [-2x,4x]',fontsize=24, color='green')

    for ii,prop in enumerate(propagators_noise):

        to_plot=[(x.imag) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot,color=colors[ii],marker=noises[ii],s=sizes[ii])

    for prop in propagators_full:
 
        to_plot=[(x.imag) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot, s=200)
    
    ylim([-2*to_plot[0],4*to_plot[0]])
    xticks([])

    subplot(3,24,t_s+49)

    if(t_s==0):
        ylabel('abs value [0.5x,1.5x]',fontsize=24, color='green')

    for ii,prop in enumerate(propagators_noise):

        to_plot=[abs(x) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot,color=colors[ii],marker=noises[ii],s=sizes[ii])

    for prop in propagators_full:
 
        to_plot=[abs(x) for x in prop[0][0][conf][t_slice]][t_s:t_s+1]
        scatter(range(len(to_plot)),to_plot, s=200)
    
    ylim([0.5*to_plot[0],1.5*to_plot[0]])
    xticks([])
    
```

## plot of differences in effective mass

```python
fig, ax = subplots(figsize=(16,20))
noises=["$1$","$4$","$10$","$20$","$50$"]
colors=["red","red","green","green","green"]
sizes=[120,120,200,200,200]

for conf in range(2):
    for t_slice in range(4):
        
        subplot(4,2,1+4*conf+t_slice)
            
        xlim([-2,25])
        ylim([6,9])

        for ii,prop in enumerate(propagators_noise):

            scatter(range(24),effective_mass(prop[0][0][conf][t_slice]),color=colors[ii],marker=noises[ii],s=sizes[ii])

        for prop in propagators_full:
 
            scatter(range(24),effective_mass(prop[0][0][conf][t_slice]),s=80)
    
```
