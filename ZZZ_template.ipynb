## plot of effective mass

```python
# Two subplots, the axes array is 1-d


p_measurements  = measurements
p_particles     = propagators_single_hadron[0][1]
p_configurations= propagators_single_hadron[0][2]

time_slices=range(int(p_configurations[0][5:7]))

fig= plt.figure(figsize=(6*len(measurements),4*len(p_particles)))

iii=0
pp=[]
for part_i, part in enumerate(p_particles):
    for meas_i, meas in enumerate(p_measurements):
        
        if(iii%3!=0):
            pp.append(plt.subplot(len(p_particles),len(p_measurements),iii+1, sharey=pp[iii-1]))
        else:            
            pp.append(plt.subplot(len(p_particles),len(p_measurements),iii+1))
        pp[iii].set_title(part+" "+meas)
            
        iii=iii+1
        for conf_i, conf in enumerate(p_configurations):
            for i in range(len(propagators_single_hadron[meas_i][0][part_i])):
                scatter(time_slices,effective_mass(propagators_single_hadron[meas_i][0][part_i][conf_i][i]))
```
